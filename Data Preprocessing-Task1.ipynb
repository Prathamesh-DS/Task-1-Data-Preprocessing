## Convert date formats to a consistent type (e.g., dd-mm-yyyy).

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)
df['Delivery Date'] = pd.to_datetime(df['Delivery Date'], errors='coerce', dayfirst=True)
df['Cancellation Date'] = pd.to_datetime(df['Cancellation Date'], errors='coerce', dayfirst=True)

# pd.to_datatime() to convert them to Pandas datetime format
# error='coerce' handles bad or missing dates by converting them to NaT(not a time)
# dayfirst=True treats 12-05-2024 as 12 may 2024
```

```python
df
```

## Rename column headers to be clean and uniform (e.g., lowercase, no spaces).

```python
df.columns = ( df.columns.str.lower().str.strip().str.replace(" ","_").str.replace("/","_"))

# converts all column names to lowercase

#.str.strip()- removes extra spaces at the beginning or end

#.str.replace(" ", "_")- replaces spaces with underscores

#.str.replace("/", "_")- replaces slashes with underscores 
```

```python
df
```

## Check and fix data types (e.g., age should be int, date as datetime).

```python
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce', dayfirst=True)
df['delivery_date'] = pd.to_datetime(df['delivery_date'], errors='coerce', dayfirst=True)
df['cancellation_date'] = pd.to_datetime(df['cancellation_date'], errors='coerce', dayfirst=True)
```

```python
df.dtypes
```

```python
df['customer_age'] = pd.to_numeric(df['customer_age'], errors='coerce')
df['mrp'] = pd.to_numeric(df['mrp'], errors='coerce')
df['shipping_charges'] = pd.to_numeric(df['shipping_charges'], errors='coerce')
```
